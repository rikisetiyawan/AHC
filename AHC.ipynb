```python
# import library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

Nama : RIKI SETIYAWAN (A11.2018.11410)

```python
data = pd.read_excel('Data AHC.xlsx')
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
# melakukan hierachical klastering complete linkage
Z = linkage(x, 'complete')
```

```python
print(Z)
```

```python
# plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Z)
plt.show()
```

```python
# melakukan hierachical klastering Average linkage
Y = linkage(x, 'average')
```

```python
print(Y)
```

```python
# plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Y)
plt.show()
```
